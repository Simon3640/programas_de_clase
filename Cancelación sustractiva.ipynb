# Evaluando una ecuación cuadratica, cancelación sustractiva

```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
a=1
b=1
x_list=[]
y_list=[]
for i in range(1,10):
    c=10**(-i)
    x_1=(-b+(b**2-4*a*c)**0.5)/(2*a)
    x_1_p= -(2*c)/(b+(b**2-4*a*c)**0.5)
    error=abs(x_1_p-x_1)/abs(x_1_p)*100.0
    x_list.append(c)
    y_list.append(error)
   # print("c=%.5e x_1=%.16e x_1_p=%.16e error=%.16f" %(c,x_1,x_1_p,error))
```

# Gráfica del error 

```python
fig=plt.figure(figsize=(6,6))
ax=plt.axes()
ax.grid()
ax.plot(x_list,y_list,'o')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel("c")
ax.set_ylabel("error")
plt.show()
```
